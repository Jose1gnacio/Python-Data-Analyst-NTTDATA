```python
#Función para calcular el indice de criminalidad por cada 100.000 habitantes.
def indice_criminalidad(datos):
  indices = []
  for dato in datos:
    num_delitos = dato["Número de delitos"]
    poblacion = dato["Población"]
    if poblacion == 0:
        continue
    indice = (num_delitos / poblacion) * 100000
    indices.append(round(indice))
  return indices

# Datos de delitos y población por área comunitaria.
datos = [
    {"Área": 1, "Número de delitos": 289, "Población": 54991},
    {"Área": 2, "Número de delitos": 228, "Población": 71942},
    {"Área": 3, "Número de delitos": 319, "Población": 56362},
    {"Área": 4, "Número de delitos": 141, "Población": 39493},
    {"Área": 5, "Número de delitos": 5, "Población": 0}
]

# Calcular el índice de criminalidad para cada área.
indices = indice_criminalidad(datos)

# Encontrar el máximo, mínimo y medio.
maximo = max(indices)
minimo = min(indices)
medio = sum(indices) / len(indices)

#Mostramos los resultados
print(indices)
print("Máximo:", maximo)
print("Mínimo:", minimo)
print("Medio:", medio)
```

```python
from datetime import datetime

#Función para calcular el tiempo en minutos entre cada delito almacenado.
def calcular_tiempo_entre_delitos(delitos):
  # Ordenar los delitos por fecha de forma ascendente
  delitos_ordenados = sorted(delitos, key=lambda x: x["Fecha"])

  # Calcular la diferencia de tiempo entre delitos consecutivos
  print("Tiempo entre delitos consecutivos:")
  for i in range(1, len(delitos_ordenados)):
    fecha_actual = datetime.strptime(delitos_ordenados[i]["Fecha"], "%Y-%m-%d %H:%M:%S")
    fecha_anterior = datetime.strptime(delitos_ordenados[i - 1]["Fecha"], "%Y-%m-%d %H:%M:%S")
    diferencia = fecha_actual - fecha_anterior
    tiempo_en_minutos = diferencia.total_seconds() / 60
    num_caso_actual = delitos_ordenados[i]["Núm caso"]
    num_caso_anterior = delitos_ordenados[i - 1]["Núm caso"]
    print(f"El tiempo entre el caso número {num_caso_anterior} y el caso número {num_caso_actual} es de {tiempo_en_minutos:.2f} minutos")

# Calcular el tiempo entre delitos cada delito
calcular_tiempo_entre_delitos(delitos)

```

```python
!pip install pyproj

from pyproj import Transformer
import numpy as np
import math

sistema_referencia_origen ="EPSG:4326"  #sistema geográfico.
sistema_referencia_destino ="EPSG:26916" #sistema proyeado.


'''
Trasnformamos las coordenadas geográficas de los delitos en coordenadas proyectadas.
'''
#Función que transforma las coordenadas geográficas a coordenadas proyectadas de los delitos.
def convertir_coordenadas_delitos(delitos, sistema_referencia_origen, sistema_referencia_destino):

  transformer = Transformer.from_crs(sistema_referencia_origen, sistema_referencia_destino)

  delitos_transformados = []

  for delito in delitos:
    coordenadas = delito["Coordenadas"]
    latitud = coordenadas["Latitud"]
    longitud = coordenadas["Longitud"]
    x, y = transformer.transform(latitud, longitud)
    delito_transformado = {
        "Núm caso": delito["Núm caso"],
        "Descripción": delito["Descripción"],
        "Arrestado": delito["Arrestado"],
        "Núm área comunitaria": delito["Núm área comunitaria"],
        "Cuadra": delito["Cuadra"],
        "Fecha": delito["Fecha"],
        "Coordenadas Geográficas": {"Latitud": latitud, "Longitud": longitud},
        "Coordenadas Proyectadas": {"X": x, "Y": y}
    }
    delitos_transformados.append(delito_transformado)

  return delitos_transformados

#Guardamos las nuevas coordenadas en la información actual en una nueva variable.
delitos_transformados = convertir_coordenadas_delitos(delitos, sistema_referencia_origen, sistema_referencia_destino)

#Mostramos los delitos con las coordenadas geográficas y proyectadas
#for delito in delitos_transformados:
#  print(f"Núm caso: {delito['Núm caso']}")
#  print(f"Coordenadas geográficas: {delito['Coordenadas Geográficas']}")
#  print(f"Coordenadas proyectadas: {delito['Coordenadas Proyectadas']}")
#  print()

################################################################################


'''
Trasnformamos las coordenadas geográficas de las comisarias en coordenadas proyectadas.
'''
#Función que transforma las coordenadas geográficas a coordenadas proyectadas de las comisarias
def convertir_coordenadas_comisarias(ubicaciones_comisarias, sistema_referencia_origen, sistema_referencia_destino):
  transformer = Transformer.from_crs(sistema_referencia_origen, sistema_referencia_destino)

  comisarias__transformadas = []

  for ubicacion in ubicaciones_comisarias:
    latitud, longitud, comisaria = ubicacion
    x, y = transformer.transform(latitud, longitud)
    comisaria__transformada = {
        "Comisaria": comisaria,
        "Coordenadas Geográficas": {
            "Latitud": latitud,
            "Longitud": longitud
        },
        "Coordenadas Proyectadas": {
            "X": x,
            "Y": y
        }
    }
    comisarias__transformadas.append(comisaria__transformada)

  return comisarias__transformadas

#Llamamos a la función
comisarias_transformadas = convertir_coordenadas_comisarias(ubicaciones_comisarias, sistema_referencia_origen, sistema_referencia_destino)

#Mostramos las comisarias con las coordenadas geográficas y proyectadas
#for comisaria in comisarias_transformadas:
#  print(f"Nombre: {comisaria['Comisaria']}")
#  print(f"Coordenadas geográficas: {comisaria['Coordenadas Geográficas']}")
#  print(f"Coordenadas proyectadas: {comisaria['Coordenadas Proyectadas']}")
#  print()

################################################################################


'''
Cálculo de comisaria mas cercana a cada delito.
'''
#Función para calcular la distancia
def calcular_distancia(punto1, punto2):
  return np.sqrt((punto1['X'] - punto2['X'])**2 + (punto1['Y'] - punto2['Y'])**2)

#Función principal para encontrar la comisaría más cercana a cada delito y guardar esa comisaria en la información del delito
def encontrar_comisaria_mas_cercana(delitos_transformados, comisarias_transformadas):
  resultados = []

  for delito in delitos_transformados:
    coordenadas_delito = delito['Coordenadas Proyectadas']
    distancia_minima = float('inf')
    comisaria_cercana = None

    for comisaria in comisarias_transformadas:
      coordenadas_comisaria = comisaria['Coordenadas Proyectadas']
      distancia = calcular_distancia(coordenadas_delito, coordenadas_comisaria)

      if distancia < distancia_minima:
        distancia_minima = distancia
        comisaria_cercana = comisaria

    if distancia_minima < 1000:
      distancia = f"{distancia_minima:.2f} metros"
    else:
      distancia = f"{distancia_minima / 1000:.2f} kilómetros"

    resultados.append({
        "Núm caso": delito["Núm caso"],
        "Descripción": delito["Descripción"],
        "Arrestado": delito["Arrestado"],
        "Núm área comunitaria": delito["Núm área comunitaria"],
        "Cuadra": delito["Cuadra"],
        "Fecha": delito["Fecha"],
        "Coordenadas Geográficas": delito["Coordenadas Geográficas"],
        "Coordenadas Proyectadas": delito["Coordenadas Proyectadas"],
        "Comisaria Más Cercana": comisaria_cercana["Comisaria"],
        "Distancia a la Comisaria Más Cercana": distancia
    })

  return resultados

#Llamamos a la función
delitos_y_comisarias = encontrar_comisaria_mas_cercana(delitos_transformados, comisarias_transformadas)

#Mostramos los delitos y la comisaria mas cercana
for delito in delitos_y_comisarias:
  print(f"Delito {delito['Núm caso']}:")
  print(f"  Descripción: {delito['Descripción']}")
  print(f"  Comisaria Más Cercana: {delito['Comisaria Más Cercana']}")
  print(f"  Distancia: {delito['Distancia a la Comisaria Más Cercana']}")

################################################################################
'''
FIN.
'''
```
